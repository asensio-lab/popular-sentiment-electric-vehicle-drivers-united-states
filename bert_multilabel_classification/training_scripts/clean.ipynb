```python
truncated = combined[combined['Value.participantid'].isin(['1', '2', '3', '4', '5'])]
```

```python
truncated = truncated.rename(columns={'Value.function':'functionality', 'Value.range':'range_anxiety', 'Value.availability':'availability', 'Value.cost':'cost', 'Value.userinteraction':'ui', 'Value.location':'location', 'Value.servicetimes':'service_time', 'Value.dealership':'dealership', 'Value.other':'other'})
```

```python
truncated_col = truncated[['id', 'review', 'functionality', 'range_anxiety', 'availability', 'cost', 'ui', 'location', 'service_time', 'dealership']]
```

```python
truncated_col.head(20)
```

```python
truncated_col.fillna(0)
```

```python
truncated_col = truncated_col.astype(str)
```

```python
truncated_col['functionality'] = truncated_col['functionality'].replace('nan', '0')
truncated_col['functionality'] = truncated_col['functionality'].replace('True', '1')
truncated_col['range_anxiety'] = truncated_col['range_anxiety'].replace('nan', '0')
truncated_col['range_anxiety'] = truncated_col['range_anxiety'].replace('True', '1')
truncated_col['availability'] = truncated_col['availability'].replace('nan', '0')
truncated_col['availability'] = truncated_col['availability'].replace('True', '1')
truncated_col['cost'] = truncated_col['cost'].replace('nan', '0')
truncated_col['cost'] = truncated_col['cost'].replace('True', '1')
truncated_col['ui'] = truncated_col['ui'].replace('nan', '0')
truncated_col['ui'] = truncated_col['ui'].replace('True', '1')
truncated_col['location'] = truncated_col['location'].replace('nan', '0')
truncated_col['location'] = truncated_col['location'].replace('True', '1')
truncated_col['service_time'] = truncated_col['service_time'].replace('nan', '0')
truncated_col['service_time'] = truncated_col['service_time'].replace('True', '1')
truncated_col['dealership'] = truncated_col['dealership'].replace('nan', '0')
truncated_col['dealership'] = truncated_col['dealership'].replace('True', '1')
```
